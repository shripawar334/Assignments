### EDA -> COVID 19


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('covid19_Confirmed_dataset.csv')
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.describe()
```

```python
df.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
df.head(20)
```

```python
corona_datase_aggregated=df.groupby('Country/Region').sum()
```

```python
corona_datase_aggregated.head(20)
```

```python
corona_datase_aggregated.loc['Benin'][:3].plot()
# corona_datase_aggregated.loc['Bahrain'].plot()
# corona_datase_aggregated.loc['Bhutan'].plot()
plt.legend()
```

### calculate first derivative of curve

```python
corona_datase_aggregated.loc['Bahrain'].diff().plot()
```

### maximum infection rate

```python
corona_datase_aggregated.loc['Bahrain'].diff().max()
```

```python
corona_datase_aggregated.loc['Bhutan'].diff().max()
```

```python
corona_datase_aggregated.loc['Austria'].diff().max()
```

```python
countries = list(corona_datase_aggregated.index)
max_infection_rates = []

for c in countries:
    # Convert all values to numeric (ignore non-numeric)
    country_data = pd.to_numeric(corona_datase_aggregated.loc[c], errors='coerce')
    
    # Find max infection rate (difference between consecutive days)
    max_infection_rates.append(country_data.diff().max())

# Add as new column
corona_datase_aggregated['max_infection_rates'] = max_infection_rates

```

```python
corona_datase_aggregated.head()
```

### create new dataframe 

```python
corona_data=pd.DataFrame(corona_datase_aggregated['max_infection_rates'])
```

```python
corona_data
```
